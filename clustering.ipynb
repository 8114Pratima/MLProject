```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# K-Means Clustering

## Import Libraries and Dataset

```python
#importing the dataset
dataset = pd.read_csv('https://drive.usercontent.google.com/download?id=1kNmYKBaSgeCa0KUvu8OP8KVZXoA6p3x6&export=download&authuser=0&confirm=t&uuid=2e951ed4-a323-44cb-a863-9b0b8af383b5&at=APZUnTW8RWZWA3xRRXrOlOQELJoH:1720609939055')
#creating a duplicate dataset to work on
dataset1 = dataset
# peeking at the dataset
dataset1.head()
```

```python
#Descriptive stats of the variables in data
dataset1.describe()
```

## Feature Scaling

```python
data = dataset1.iloc[:, 2:7].values  #standardize the data to normal distribution
from sklearn import preprocessing
dataset1_standardized = preprocessing.scale(data)
dataset1_standardized = pd.DataFrame(dataset1_standardized)
print (dataset1_standardized)
```

## WSS Plot or Elbow Curve for Optimal cluster identification

```python
# find the appropriate cluster number
plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(dataset1_standardized)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## K-Means Clustering & Cluster Information

```python
from sklearn.cluster import KMeans
import pandas as pd

# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(dataset1_standardized)

# Beginning of the cluster numbering with 1 instead of 0
y_kmeans1 = y_kmeans + 1

# New DataFrame called cluster
cluster = pd.DataFrame(y_kmeans1, columns=['cluster'])

# Adding cluster to the dataset1
dataset1['cluster'] = cluster

# Selecting only numeric columns for mean calculation
numeric_columns = dataset1.select_dtypes(include=['number'])

# Adding the cluster column back to the numeric columns DataFrame
numeric_columns['cluster'] = dataset1['cluster']

# Mean of clusters
kmeans_mean_cluster = pd.DataFrame(round(numeric_columns.groupby('cluster').mean(), 1))

# Display the mean of clusters
print(kmeans_mean_cluster)
```
